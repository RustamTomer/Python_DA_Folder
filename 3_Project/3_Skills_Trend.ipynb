```python
from datasets import load_dataset
import pandas as pd
import matplotlib.pyplot as plt
import ast
import seaborn as sns
import adjustText as adjust_text

dataset = load_dataset('lukebarousse/data_jobs')

df = dataset['train'].to_pandas()

df['job_posted_date'] = pd.to_datetime(df['job_posted_date'])
df['job_skills'] = df['job_skills'].apply(lambda x : ast.literal_eval(x) if pd.notna(x) else x)
```

```python
df_DA_US = df[(df['job_title_short'] == 'Data Analyst') & (df['job_country'] == 'United States')].copy()

df_DA_US['job_posted_month_no'] = df_DA_US['job_posted_date'].dt.month

df_DA_US_explode = df_DA_US.explode('job_skills')
```

```python
df_DA_US_pivot = df_DA_US_explode.pivot_table(index='job_posted_month_no', columns='job_skills', aggfunc='size', fill_value=0)

df_DA_US_pivot.loc['Total'] = df_DA_US_pivot.sum()
df_DA_US_pivot = df_DA_US_pivot[df_DA_US_pivot.loc['Total'].sort_values(ascending=False).index]
df_DA_US_pivot.drop('Total', inplace=True)
```

```python
df_DA_US_mat = df_DA_US_pivot.iloc[:11, :5].plot(kind='line')
plt.title('Top 5 Skills for Data Analyst per Month')
plt.xlabel('')
plt.ylabel('Count of Job Postings')

from matplotlib.ticker import PercentFormatter

ax=plt.gca()
ax.yaxis.set_major_formatter(PercentFormatter(decimals=0))
plt.show()
```

```python
DA_total = df_DA_US.groupby('job_posted_month_no').size()
```

```python
df_DA_US_pivot = df_DA_US_pivot.div(DA_total/100, axis=0)
df_DA_US_pivot = df_DA_US_pivot.reset_index()
df_DA_US_pivot['job_posted_month'] = df_DA_US_pivot['job_posted_month_no'].apply(lambda x : pd.to_datetime(x, format='%m').strftime('%b'))
df_DA_US_pivot.set_index('job_posted_month', inplace=True)
df_DA_US_pivot = df_DA_US_pivot.drop(columns='job_posted_month_no')
df_DA_US_pivot
```

```python
df_plot = df_DA_US_pivot.iloc[:11, :5]

sns.set_theme(style='ticks')
sns.lineplot(data=df_plot, dashes=False, palette='tab10')
sns.despine()

for i in range(5):
    plt.text(10.2, df_plot.iloc[-1, i], df_plot.columns[i])
plt.title('Trending Top Skills for Data Analysts in the US')
plt.xlabel('2023')
plt.ylabel('Likelihood of Job Postings')
plt.legend().remove()

plt.show()
```
