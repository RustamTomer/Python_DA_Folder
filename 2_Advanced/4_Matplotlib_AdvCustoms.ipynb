```python
from datasets import load_dataset
import pandas as pd
import matplotlib.pyplot as plt
import ast
import adjustText as adjust_text

dataset = load_dataset('lukebarousse/data_jobs')

df = dataset['train'].to_pandas()

df['job_posted_date'] = pd.to_datetime(df['job_posted_date'])
df['job_skills'] = df['job_skills'].apply(lambda x : ast.literal_eval(x) if pd.notna(x) else x)
df_DA = df[df['job_title_short'] == 'Data Analyst'].copy()
```

```python
df_DA['job_posted_month_no']=df['job_posted_date'].dt.month

df_DA_exploded = df_DA.explode('job_skills')
df_DA_pivot = df_DA_exploded.pivot_table(index='job_posted_month_no', columns='job_skills', aggfunc='size', fill_value=0)
df_DA_pivot.loc['Total'] = df_DA_pivot.sum()
df_DA_pivot = df_DA_pivot[df_DA_pivot.loc['Total'].sort_values(ascending=False).index]
df_DA_pivot = df_DA_pivot.drop('Total')

df_DA_pivot.reset_index(inplace=True)
df_DA_pivot['job_posted_month'] = df_DA_pivot['job_posted_month_no'].apply(lambda x : pd.to_datetime(x, format='%m').strftime('%b'))
df_DA_pivot.set_index('job_posted_month', inplace=True)
df_DA_pivot = df_DA_pivot.drop(columns='job_posted_month_no')
df_DA_pivot
```

```python
df_DA_pivot.iloc[:, :5].plot(
    kind='line',
    linewidth = 4,
    linestyle = ':',
    colormap= 'viridis',
    marker = 'o',
    markersize = 5,
    figsize=(10,5)
)

plt.title('Top 5 Skills for Data Analyst per Month')
plt.xlabel('')
plt.ylabel('Count f Job Postings')
plt.show()
```

```python
df_DA_skills = df_DA.explode('job_skills')
skills_count = df_DA_skills.groupby('job_skills').agg(
    skill_count = ('job_skills', 'count'),
    median_salary = ('salary_year_avg', 'median')
)

filter_count = 20
skills_count = skills_count.sort_values(by='skill_count', ascending=False).head(filter_count)
```

```python
skills_count.plot(kind = 'scatter', x='skill_count', y='median_salary')


for i, txt in enumerate(skills_count.index):
    (plt.text(skills_count['skill_count'].iloc[i], skills_count['median_salary'].iloc[i], txt))

ax = plt.gca()
ax.yaxis.set_major_formatter(plt.FuncFormatter(lambda y, pos:f'${int(y/1000)}K'))

plt.title(f'Salary vs. Count of Job Postings for Top {filter_count} Skills')
plt.ylabel('Median Yearly Salary ($USD)')
plt.xlabel('Count of Job Postings')
plt.tight_layout()
plt.show()
```
