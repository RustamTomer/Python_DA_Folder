```python
!pip install datasets
from datasets import load_dataset
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

dataset = load_dataset('lukebarousse/data_jobs')

df = dataset['train'].to_pandas()

df['job_posted_date'] = pd.to_datetime(df['job_posted_date'])
```

```python
job_counts = df['job_title_short'].value_counts()

plt.bar(job_counts.index, job_counts.values)
plt.show()
```

```python
plt.bar(job_counts.index, job_counts.values)
plt.title("Postings by Job Title")
plt.ylabel('Count of Job Postings')
plt.xticks(rotation=45, ha='right')
plt.show()
```

```python
job_counts.plot(kind='bar')
plt.title("Postings by Job Title")
plt.ylabel("Count of Job Postings")
plt.xlabel("")
plt.xticks(rotation = 45, ha='right')
plt.show()
```

```python
df['job_posted_month'] = df['job_posted_date'].dt.month
df.plot(x='job_posted_month', y='salary_year_avg', kind='bar')
```

EXERCISE : MATPLOTLIB BASICS

```python
jobs_salary = df.groupby('job_title_short')['salary_year_avg'].median().sort_values()

jobs_salary.plot(kind='barh')
plt.title('Median Salary by Job Title')
plt.xlabel("Salary ($USD)")
plt.ylabel(' ')
plt.show()
```
